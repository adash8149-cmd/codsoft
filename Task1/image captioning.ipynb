```python
import os
from PIL import Image
from transformers import BlipProcessor, BlipForConditionalGeneration

def generate_caption(image_path):
    print("рџ”„ Loading AI Model (this takes a moment on the first run)...")
    processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
    model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base")
    
    if not os.path.exists(image_path):
        return f"Error: Image '{image_path}' not found. Please check the file name."

    raw_image = Image.open(image_path).convert('RGB')
    
    # Display the image inside Jupyter Notebook
    display(raw_image)
    
    print("рџ“ё Analyzing image...")
    inputs = processor(raw_image, return_tensors="pt")
    out = model.generate(**inputs, max_new_tokens=50)
    caption = processor.decode(out[0], skip_special_tokens=True)
    return caption

# Run the function (Make sure to save a picture named 'test.jpg' in your notebook folder)
img_path = "test.jpg" 
try:
    caption = generate_caption(img_path)
    print(f"\nGenerated Caption: {caption.capitalize()}")
except Exception as e:
    print(f"An error occurred: {e}")

```
